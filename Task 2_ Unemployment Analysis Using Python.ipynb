# Data Cleaning & Preparation

```python
print(df.columns)
```

```python
# Remove leading/trailing spaces from all column names
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)
```

```python
# Rename columns for ease
df = df.rename(columns={
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate',
    'Frequency': 'Month'
})
```

```python
df.isnull().sum()
```

```python
# Fill missing values
df.fillna(0, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Unemployment Rate Over Time
plt.figure(figsize=(8,6))
plt.plot(df['Date'], df['Unemployment_Rate'])
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Convert Date column
df.columns = df.columns.str.strip()

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

print(df['Date'].dtype)
```

```python
# Split data
covid_start = pd.to_datetime("2020-03-01")

pre_covid = df[df['Date'] < covid_start]
covid = df[df['Date'] >= covid_start]
```

```python
# Covid-19 Impact Analysis
plt.figure(figsize=(12,6))
plt.plot(pre_covid['Date'], pre_covid['Unemployment_Rate'], label='Pre-Covid')
plt.plot(covid['Date'], covid['Unemployment_Rate'], label='Covid Period', color='red')
plt.legend()
plt.title("Impact of Covid-19 on Unemployment")
plt.show()
```

```python
df['Month'] = df['Date'].dt.month
```

```python
# Monthly average unemployment
monthly_avg = df.groupby('Month')['Unemployment_Rate'].mean()

plt.figure(figsize=(8,5))
monthly_avg.plot(kind='bar')
plt.title("Average Monthly Unemployment Rate")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Rural vs Urban Comparison
plt.figure(figsize=(8,5))
sns.boxplot(x='Area', y='Unemployment_Rate', data=df)
plt.title("Unemployment Rate: Rural vs Urban")
plt.show()
```

```python
# Distribution of Region
plt.figure(figsize=(15,8))
sns.countplot(x='Region', data=df)
plt.title('Distribution of Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Summary
print(f"\nSummary of Region:\n")
print(df['Region'].value_counts())
print("\nPercentage Distribution:\n")
print((df['Region'].value_counts(normalize=True) * 100).round(2))
```

### Bivariate Analysis

```python
df.head()
```

```python
# Average unemployment rate by region
region_summary = df.groupby('Region')['Unemployment_Rate'].agg(
  Count='count', Mean='mean',Median='median',Std='std',Min='min',Max='max'
).sort_values(by='Mean', ascending=False)

# Plotting
plt.figure(figsize=(12,6))
sns.barplot(
    x=region_summary.index,
    y=region_summary['Mean'],
    palette='viridis'
)
plt.xticks(rotation=90)
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate')
plt.show()

# Summary
print("\nRegion-wise Unemployment Rate Summary:\n")
print(region_summary.round(2))
```

```python
# Average unemployment rate by Month
Month_summary = df.groupby('Month')['Unemployment_Rate'].agg(
  Count='count', Mean='mean',Median='median',Std='std',Min='min',Max='max'
).sort_values(by='Mean', ascending=False)

# Plotting
plt.figure(figsize=(7,5))
sns.barplot(
    x=Month_summary.index,
    y=Month_summary['Mean'],
    palette='viridis'
)
plt.xticks(rotation=90)
plt.title('Average Estimated Unemployment Rate by Month')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.show()

# Summary
print("\nFrequency-wise Unemployment Rate Summary:\n")
print(Month_summary.round(2))
```

```python
# Average unemployment rate by Area
Area_summary = df.groupby('Area')['Unemployment_Rate'].agg(
  Count='count', Mean='mean',Median='median',Std='std',Min='min',Max='max'
).sort_values(by='Mean', ascending=False)

# Plotting
plt.figure(figsize=(7,5))
sns.barplot(
    x=Area_summary.index,
    y=Area_summary['Mean'],
    palette='viridis'
)
plt.xticks(rotation=90)
plt.title('Average Estimated Unemployment Rate by Area')
plt.xlabel('Area')
plt.ylabel('Unemployment Rate (%)')
plt.show()

# Summary
print("\nArea-wise Unemployment Rate Summary:\n")
print(Area_summary.round(2))
```

```python
# Average unemployment for each labour participation
plt.figure(figsize=(10,5))
sns.lineplot(
    data=df,
    x='Labour_Participation_Rate',
    y='Unemployment_Rate',
    marker='o'
)
plt.title('Unemployment Rate vs Labour Participation Rate')
plt.xlabel('Labour Participation')
plt.ylabel('Unemployment Rate ')
plt.show()

# Summary AFTER graph
print(f"\nStatistical Summary of Labour Participation:\n")
print(df['Labour_Participation_Rate'].describe().round(2))

print(f"\nStatistical Summary of Unemployment Rate:\n")
print(df['Unemployment_Rate'].describe().round(2))
```

```python
# Average unemployment for each Estimated Employed
plt.figure(figsize=(10,5))
sns.lineplot(
    data=df,
    x='Estimated Employed',
    y='Unemployment_Rate',
    marker='o'
)
plt.title('Unemployment Rate vs Estimated EmployedRate')
plt.xlabel('Estimated Employed')
plt.ylabel('Unemployment Rate ')
plt.show()

# Summary
print(f"\nStatistical Summary of Estimated Employed:\n")
print(df['Estimated Employed'].describe().round(2))

print(f"\nStatistical Summary of Unemployment Rate:\n")
print(df['Unemployment_Rate'].describe().round(2))
```

```python
numeric_cols = ['Unemployment_Rate', 'Estimated Employed', 'Labour_Participation_Rate']

# Correlation matrix
corr = df[numeric_cols].corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Columns")
plt.show()
```
